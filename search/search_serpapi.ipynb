```python
import serpapi
import os
from dotenv import load_dotenv
```

```python
load_dotenv()
params = {
  "engine": "google_scholar",
  "q": "foot and mouth disease control",
  "as_ylo": "2025",
  "api_key": f"{os.getenv('SERPAPI_API_KEY')}"
}
```

```python
search = serpapi.search(params)
items = search["organic_results"]
```

```python

for item in items:
    title = item.get('title', '')
    url = item.get('link', '')

    print(f'Title: {title}')
    print(f'URL: {url}\n')
```
