## Semantic Scholar
https://www.semanticscholar.org/product/api/tutorial

```python
import requests
import os
from load_dotenv import load_dotenv
```

```python
load_dotenv()
base_url = "https://api.semanticscholar.org/graph/v1/paper/search"
headers = {
    "x-api-key": os.getenv("SEMANTIC_SCHOLAR_API_KEY")
}
```

```python
# https://www.semanticscholar.org/product/api/tutorial#search-query-params
query = "foot and mouth disease in Canada" 

query_params = {
    "query": query,
    "fields": "title,url,abstract,year",
    "year": "2024-"
}

response = requests.get(base_url, params=query_params, headers=headers).json()
```

```python
articles = response.get("data", [])
for article in articles:
    print(f"Title: {article['title']}")
    print(f"URL: {article['url']}")
    print(" ")
```
