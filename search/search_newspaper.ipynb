## Newspaper
Building a news source   
https://newspaper.readthedocs.io/en/latest/user_guide/quickstart.html#building-a-news-source

```python
import newspaper
import nltk
```

```python
#nltk.download('punkt')
source_path = "https://animalhealthcanada.ca/news"
items = newspaper.build(source_path, memoize_articles=False)
```

```python
for item in items.articles:

    item.download()
    item.parse()
    item.nlp()

    title = item.title
    url = item.url

    print(f"Title: {title}\nURL: {url}\n")
```
