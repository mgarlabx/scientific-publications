## Web Search


```python
import os
from load_dotenv import load_dotenv
from openai import OpenAI
```

```python
input = """
You are a helpful assistant that reads news about foot-and-mouth disease in Canada.
Find the most recent (max 3 months) information about foot-and-mouth disease outbreaks, control measures, and government responses.
Present the results in this format:
Title: <title>
URL: <url>
"""
```

```python
load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
response = client.responses.create(
    model="gpt-5-nano",
    input=input,
    tools=[{"type": "web_search"}],
)
```

```python
print(response.output_text)
```
