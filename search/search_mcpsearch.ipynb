## MCP Search
https://smithery.ai/server/@lecigarevolant/arxiv-mcp-server-gpt


```python
import os
from load_dotenv import load_dotenv
from openai import OpenAI
```

```python
input = """
You are a helpful assistant that reads articles about foot-and-mouth disease in any part of the world.
Find the most recent (max 12 months) information about foot-and-mouth disease outbreaks, control measures, and government responses.
Use the MCP tool to search for relevant articles.
Present the results in this format:
Title: <title>
URL: <url>
"""
```

```python
load_dotenv()
client = OpenAI(api_key=os.getenv("OPENAI_API_KEY"))
response = client.responses.create(
    model="gpt-5-nano",
    input=input,
    tools=[
        {
            "type": "mcp",
            "server_label": "smithery",
            "server_url": f"https://server.smithery.ai/@lecigarevolant/arxiv-mcp-server-gpt/mcp?api_key={os.getenv('SMITHERY_API_KEY')}",
            "require_approval": "never"
        }
    ],
)
```
