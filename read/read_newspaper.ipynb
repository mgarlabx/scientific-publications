## Newspaper
Get an article from URL     
https://newspaper.readthedocs.io/en/latest/user_guide/quickstart.html#news-articles

```python
from newspaper import Article
```

```python
url = "https://infodemiology.jmir.org/2025/1/e75434"
```

```python
article = Article(url)
article.download()
article.parse()
article.nlp()
```
