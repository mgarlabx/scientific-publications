## Crossref
A non-for-profit organisation based in USA focused on to make research outputs easier to find, cite, link, assess, and reuse.    
https://www.crossref.org/about


```python
from habanero import Crossref # https://pypi.org/project/habanero/
from bs4 import BeautifulSoup
```

```python
doi = "10.2196/75434"
```

```python
cr = Crossref()
work = cr.works(ids=doi)
article = work['message']
```

```python
article["title"][0]
```

```python
article['URL']
```

```python
# JATS = Journal Article Tag Suite - https://www.niso.org/standards-committees/jats
article['abstract']
```

```python
abstract_cleaned = BeautifulSoup(article['abstract'], 'lxml').get_text()
```

```python
abstract_cleaned
```

```python
for author in article["author"]:
    print(f"{author['given']} {author['family']}")
    print("-"*80)
```

```python
for reference in article["reference"]:
    print(reference.get("DOI", "No DOI info"))
    print(reference.get("article-title", "No title info"))
    print("-"*80)
```
